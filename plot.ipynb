```python
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline, BSpline
import numpy as np
```

```python
n_topics = np.array([10,20,30,40,50])
xnew = np.linspace(n_topics.min(),n_topics.max(),15)
```

```python
lda60 = np.array([679.7570, 665.8643, 633.1215, 641.7270, 699.5496])
btm60 = np.array([600.4049, 543.2972, 583.5609, 586.7100, 612.52])
ptm60 = np.array([742.1569, 702.4521, 752.6027, 763.1491, 736.6593])
```

```python
lda_p = make_interp_spline(n_topics, lda60, k=3)
lda = lda_p(xnew)

btm_p = make_interp_spline(n_topics, btm60, k=3)
btm = btm_p(xnew)

ptm_p = make_interp_spline(n_topics, ptm60, k=3)
ptm = ptm_p(xnew)

plt.plot(xnew, lda, label='LDA')
plt.plot(xnew, btm, label='BTM')
plt.plot(xnew, ptm, label='PTM')
plt.legend(loc='upper left')
plt.xlabel('Topic Number')
plt.ylabel('Perplexity')
plt.ylim(500, 1000)
plt.savefig('ourstep_perplexity.png', dpi=300)
plt.show()
```
