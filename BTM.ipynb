## Instruction

This is the BTM implementation.

<b>num_t</b> specify the topics number.

<b>inferred_topics_path</b> is the file contains the documents with their corresponding inferred topic number by the BTM

<b>topics_file_path</b> is the file containing words describing topics.


```python
num_t = 40
corpus_path = '../zhihu_done.txt'
inferred_topics_path = '../btm_'+str(num_t)+'.txt'
topics_file_path = '../btm_'+str(num_t)+'_topics.txt'
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from biterm.utility import vec_to_biterms
import numpy as np
from biterm.btm import oBTM
```

```python
texts = open(corpus_path).read().splitlines()
vec = CountVectorizer()
X = vec.fit_transform(texts).toarray()
```

```python
vocab = np.array(vec.get_feature_names())
```

```python
biterms = vec_to_biterms(X)
```

```python
from biterm.btm import oBTM

btm = oBTM(num_topics=num_t, V=vocab)
topics = btm.fit_transform(biterms, iterations=50)
print("\n\n Texts & Topics ..")
with open(inferred_topics_path,'w') as btm_res:
    for i in range(len(texts)):
        topic_num = topics[i].argmax()
        print("{} (topic: {}, {})".format(texts[i], topic_num, topics[i][topic_num]))
        btm_res.write("{} (topic: {}, {})\n".format(texts[i], topic_num, topics[i][topic_num]))
        
from biterm.utility import vec_to_biterms, topic_summuary
import json

a = topic_summuary(btm.phi_wz.T, X, vocab, 20)

with open(topics_file_path,'w') as topicz:
    for i in range(20):
        topicz.write('Topic {} |  Top words= {}\n'.format(i, ' '.join(a['top_words'][i])))
```
